# Data Loading and Preprocessing

```python
# Load locally saved copy of the OECD Better Life Index excel into a pandas dataframe
# Skip the first 7 rows as these contain headers with metadata and other unnecessary information.
# Also skip the last row as this is just information on when the data was obtained,
df = pd.read_excel("OECD_betterLifeIndex.xlsx", skiprows=7, skipfooter=1)
```

```python
# As the second and third columns are just used to reference Non-OECD countries; Brzail, Russia and South Africa. This column can be removed
# and aforementioned countries can have their names added to the first column.
modified_df = df.drop(df.columns[[1, 2]], axis = 1)
```

```python
# Assign Brazil, Russia, South Africa country names in column for their rows.
brazil_row = 39
russia_row = 40
safrica_row = 41
modified_df.iloc[brazil_row, 0] = "Brazil"
modified_df.iloc[russia_row, 0] = "Russia"
modified_df.iloc[safrica_row, 0] = "South Africa"
```

```python
# Create MultiIndex headers so that each column can have a heading denoting unit, indicator, and indicator group.
# For example:
# Housing Expenditure is the indicator
# Belonging to indicator group 'Housing' (another indicator in this group would be 'Rooms per person')
# Measured in Unit 'Percentage'
indicator_groups = ['Housing', 'Income', 'Jobs', 'Community', 'Education', 'Environment', 'Civic Engagement', 'Health', 'Life Satisfaction', 'Safety', 'Work-Life Balance']
indicators = ['Dwellings w/o basic facilities',
             'Housing expenditure',
             'Rooms per person',
             'Household net adjusted disposable income',
             'Household net wealth',
             'Labour market insecurity',
             'Employment rate',
             'Long-term unemployment rate',
             'Personal earnings',
             'Quality of support network',
             'Educational attainment',
             'Student skills',
             'Years in education',
             'Air pollution',
             'Water quality',
             'Stakeholder engagement for developing regulations',
             'Voter turnout',
             'Life expectancy',
             'Self-reported health',
             'Life satisfaction',
             'Feeling safe walking alone at night',
             'Homicide rate',
             'Employees working very long hours',
             'Leisure and personal care time']
units = ['Percentage', 'Ratio', 'US Dollar', 'Average Score', 'Micrograms per m^3', 'Hours', 'Years']

# Dictionary mapping the unit to each indicator
unit_mapping = {
    'Percentage': ['Dwellings w/o basic facilities', 'Housing expenditure', 'Labour market insecurity', 'Employment rate', 'Long-term unemployment rate', 'Quality of support network', 'Educational attainment', 'Voter turnout', 'Feeling safe walking alone at night', 'Employees working very long hours', 'Water quality', 'Self-reported health'],
    'Ratio': ['Rooms per person', 'Homicide rate'],
    'US Dollar': ['Household net adjusted disposable income', 'Household net wealth', 'Personal earnings'],
    'Average Score': ['Stakeholder engagement for developing regulations', 'Life satisfaction', 'Student skills'],
    'Micrograms per m^3': ['Air pollution'],
    'Hours': ['Leisure and personal care time'],
    'Years' : ['Years in education', 'Life expectancy']
}

group_mapping = {
    'Housing' : ['Dwellings w/o basic facilities',
                 'Housing expenditure',
                 'Rooms per person'],
    'Income' : ['Household net adjusted disposable income',
                'Household net wealth'],
    'Jobs' : ['Labour market insecurity',
             'Employment rate',
             'Long-term unemployment rate',
             'Personal earnings'],
    'Community' : ['Quality of support network'],
    'Education' : ['Educational attainment',
                   'Student skills',
                   'Years in education'],
    'Environment' : ['Air pollution',
                     'Water quality'],
    'Civic Engagement' : ['Stakeholder engagement for developing regulations',
                          'Voter turnout'],
    'Health' : ['Life expectancy',
               'Self-reported health'],
    'Life Satisfaction' : ['Life satisfaction'],
    'Safety' : ['Feeling safe walking alone at night',
                'Homicide rate'],
    'Work-Life Balance' : ['Employees working very long hours',
                           'Leisure and personal care time']
}

# Create tuples which will be the MultiIndex headers
headers = []
for group, indicators in group_mapping.items():
    for indicator in indicators:
        for unit, list_of_indicators in unit_mapping.items():
            if indicator in list_of_indicators:
                headers.append((group, indicator, unit))
# Create 'multi-level' header for countries column
country_header = [('Country', '', '')]
# Create 'multi-level' header for colours column
colour_header = [('Colour', '', '')]
# Create colour column
modified_df['Colour'] = None
# Create MultiIndex from the tuples and country header
multi_index = pd.MultiIndex.from_tuples((country_header + headers + colour_header), names=['Group', 'Indicator', 'Unit'])

## Assign MuliIntext to modified dataframe
modified_df.columns = multi_index
modified_df.head()
```

```python
# Rename some country names to match with countries_shapes for the geodata
modified_df['Country'] = modified_df['Country'].replace({'Czech Republic':'Czechia','Korea':'South Korea','Slovak Republic':'Slovakia','Türkiye':'Turkey','United States':'United States of America'})

# Drop OECD row
modified_df = modified_df.drop(38)
```

```python
# Save modified df to a new excel file
modified_df.to_excel("wrangled_df.xlsx")
```

```python
# For now, use random imputation to populate empty fields so that the visualization can be worked on
for column in modified_df.columns:
    # Replace '...' being used to denote empty fields with numpy NaN value
    modified_df[column] = modified_df[column].replace('..', np.nan)

    # For each column check to see if there are any empty fields
    if modified_df[column].isna().any():
        # List of existing values
        existing_values = modified_df[column].dropna().values
        # Dataframe of rows where values are missing
        missing_rows = modified_df[column].isna()
        # Randomly select existing values to populate missing rows
        modified_df.loc[missing_rows, column] = np.random.choice(existing_values, size=missing_rows.sum())
```

```python
# Replace NaN values that have filled empty cells back to empty string for Excel, as not doing this causes some errors such as adding new rows of ony NaN values.
modified_df = modified_df.replace(np.nan, '', regex=True)

# Save the wrangled dataframe with randomly imputated values as an excel file to be loaded on and worked with in another notebook
modified_df.to_excel("wrangled_df_ri.xlsx")
```
